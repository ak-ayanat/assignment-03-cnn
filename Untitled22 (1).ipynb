```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 1. DATA
# compose barin qosady
train_transform = transforms.Compose([
    transforms.RandomRotation(10), # gradusqa bolu
    transforms.ToTensor(), # massiv
    transforms.Normalize((0.5,), (0.5,)) # -1 men 1 arasyndagy mander
])

test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

trainset = torchvision.datasets.MNIST("./data", train=True, download=True, transform=train_transform)
testset = torchvision.datasets.MNIST("./data", train=False, download=True, transform=test_transform)

trainloader = DataLoader(trainset, batch_size=64, shuffle=True) # batch 60000 nan 64-64-64 ten datany alyp train jasaidy
testloader = DataLoader(testset, batch_size=64, shuffle=False)  # shuffle kartinalardy auystyrady
```

```python
# 2. MODEL

class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__() # input 1 28 28
        self.conv1 = nn.Conv2d(1, 16, 3, padding=1) # 1-channel (rgb-3 channel)  16-16 filter arqyly 16 FM shygarady 3-kernel siza (filter)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc = nn.Linear(32 * 7 * 7, 10) # 32 channel 7x7 ar kanal olshemi

    def forward(self, x, show_features=False): # нәтижелерді қайтару
        feats = {} # әр қабаттың output-ын сақтаймыз

        x = self.conv1(x)
        feats["conv1"] = x
        x = F.relu(x)
        feats["relu1"] = x
        x = self.pool(x)
        feats["pool1"] = x

        x = self.conv2(x)
        feats["conv2"] = x
        x = F.relu(x)
        feats["relu2"] = x
        x = self.pool(x)
        feats["pool2"] = x

        x = x.view(x.size(0), -1) # бір ұзын векторға айналдырады (flatten)
        out = self.fc(x)

        return (out, feats) if show_features else out


model = SimpleCNN().to(device)
criterion = nn.CrossEntropyLoss() # қателікті есептейтін функция
optimizer = optim.Adam(model.parameters(), lr=0.001)# Оптимизатор — модельдің салмақтарын (weights) жаңартады depend on gradient

images, labels = next(iter(testloader))
img = images[0:1].to(device)

model.eval()
with torch.no_grad():
    out, feats = model(img, show_features=True)

for name, value in feats.items():
    print(name, value.shape)
```

```python
# 3. TRAINING

train_losses = []
train_accs = []
test_accs = []

for epoch in range(5):
    model.train()
    correct, total, loss_sum = 0, 0, 0

    for images, labels in trainloader:
        images, labels = images.to(device), labels.to(device) # labels дұрыс жауаптар

        optimizer.zero_grad() # gradient eseptemeidi
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward() # gradient есептеледі
        optimizer.step()

        loss_sum += loss.item()
        _, pred = torch.max(outputs, 1)
        total += labels.size(0) # 64 suret sany
        correct += (pred == labels).sum().item()

    train_acc = 100 * correct / total
    test_correct, test_total = 0, 0

    model.eval()
    with torch.no_grad():
        for images, labels in testloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, pred = torch.max(outputs, 1)

            test_total += labels.size(0)
            test_correct += (pred == labels).sum().item()

    test_acc = 100 * test_correct / test_total

    train_losses.append(loss_sum / len(trainloader))
    train_accs.append(train_acc)
    test_accs.append(test_acc)

    print(f"Epoch {epoch+1} | Loss: {train_losses[-1]:.4f} | Train: {train_acc:.2f}% | Test: {test_acc:.2f}%")
```

```python
# 4. PLOTS
plt.plot(train_losses)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

plt.plot(train_accs, label="Train")
plt.plot(test_accs, label="Test")
plt.legend()
plt.title("Accuracy")
plt.show()
```

```python
# 5. FEATURE MAPS
def show_maps(x, title, n=6):
    x = x.detach().cpu() # gradient-ті алып тастайды (графтан ажыратады)
    n = min(n, x.shape[1]) # x.shape number of channels

    fig, axes = plt.subplots(1, n, figsize=(12, 2))
    fig.suptitle(title)

    for i in range(n):
        axes[i].imshow(x[0, i], cmap="gray") # 0 бірінші batch i - n channel; imshow суретке айналдырады
        axes[i].axis("off") # осьтерді (сан, координат) өшіреді

    plt.show()

# Random test image
images, labels = next(iter(testloader))
img = images[0:1].to(device)

model.eval()
with torch.no_grad():
    out, feats = model(img, show_features=True)
    pred = out.argmax(1).item()

# Show input
img_show = img.cpu().squeeze() * 0.5 + 0.5 # артық өлшемдерді алып тастайды
plt.imshow(img_show, cmap="gray") # cmap white/black
plt.title(f"Predicted: {pred}")
plt.axis("off")
plt.show()

# Show feature maps
show_maps(feats["conv1"], "Conv1")
show_maps(feats["relu1"], "ReLU1")
show_maps(feats["pool1"], "Pool1")
show_maps(feats["conv2"], "Conv2")
show_maps(feats["relu2"], "ReLU2")
show_maps(feats["pool2"], "Pool2")
```

```python
# 6. KERNEL VISUALIZATION
def show_kernels(layer, n=6):
    weights = layer.weight.detach().cpu()

    fig, axes = plt.subplots(1, n, figsize=(12, 2))
    fig.suptitle("Conv1 Kernels")

    for i in range(n):
        axes[i].imshow(weights[i, 0], cmap="gray")
        axes[i].axis("off")

    plt.show()

show_kernels(model.conv1)
```

```python
# 7. MISCLASSIFIED IMAGES
misclassified = []

model.eval()
with torch.no_grad():
    for images, labels in testloader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        preds = outputs.argmax(1)

        for i in range(len(labels)):
            if preds[i] != labels[i]:
                misclassified.append((images[i].cpu(), labels[i].item(), preds[i].item()))
            if len(misclassified) >= 6:
                break
        if len(misclassified) >= 6:
            break

# Show misclassified
fig, axes = plt.subplots(2, 3, figsize=(8, 6))
fig.suptitle("Misclassified Images")

for ax, (img, t, p) in zip(axes.flatten(), misclassified):
    ax.imshow(img.squeeze(), cmap="gray")
    ax.set_title(f"T:{t} P:{p}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```
